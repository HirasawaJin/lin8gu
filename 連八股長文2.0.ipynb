```python
word_limit = 10000

out = []
word_count = 0
paragraph_count = 0
while word_count < word_limit:
    this_sentence = rand_from(所有句型)()
    this_sentence = " ".join(x for x in this_sentence if x)
    word_count += len(this_sentence)
    paragraph_count += len(this_sentence)
    out.append(this_sentence)
    if paragraph_count > 100:
        out.append("\n")
        paragraph_count = 0


print(" ".join(out))
```
